```python
import numpy as np
import matplotlib.pyplot as plt

# Function to find peaks and minima using sign changes
def find_peaks_and_minima(data):
    peaks = []
    minima = []
    sign = 1  # Initial sign (positive)

    for i in range(1, len(data)):
        prev_sign = sign
        sign = np.sign(data[i] - data[i - 1])

        # Identify sign change for maxima (peak)
        if prev_sign > 0 and sign < 0:
            peaks.append(i)

        # Identify sign change for minima
        elif prev_sign < 0 and sign > 0:
            minima.append(i)

    return peaks, minima

# Load data from file
data_1 = np.loadtxt('Data_1.txt')
data_2 = np.loadtxt('Data_2.txt')

# Find peaks and minima
peak_indices_1, minima_indices_1 = find_peaks_and_minima(data_1)
peak_indices_2, minima_indices_2 = find_peaks_and_minima(data_2)

# Plot Data 1 with peaks and minima
plt.figure(figsize=(10, 5))
plt.plot(data_1, label='Data 1')
plt.scatter(peak_indices_1, data_1[peak_indices_1], color='red', label='Peaks', marker='o')
plt.scatter(minima_indices_1, data_1[minima_indices_1], color='blue', label='Minima', marker='o')
plt.title('Data 1 - Sinusoidal Wave with Peaks and Minima')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()

# Plot Data 2 with peaks and minima
plt.figure(figsize=(10, 5))
plt.plot(data_2, label='Data 2')
plt.scatter(peak_indices_2, data_2[peak_indices_2], color='red', label='Peaks', marker='o')
plt.scatter(minima_indices_2, data_2[minima_indices_2], color='blue', label='Minima', marker='o')
plt.title('Data 2 - Sinusoidal Wave with Peaks and Minima')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()

# Print the indices of peaks and minima
print("Data 1 Peak indices:", peak_indices_1)
print("Data 1 Minima indices:", minima_indices_1)
print("Data 2 Peak indices:", peak_indices_2)
print("Data 2 Minima indices:", minima_indices_2)
```
